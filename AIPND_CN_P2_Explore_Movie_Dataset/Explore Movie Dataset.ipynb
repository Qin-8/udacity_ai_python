**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
df = movie_data_clean.copy()
plt.figure(figsize = [50,40])
df['net_profit'] = df['revenue'] - df['budget']
sns.regplot(data = df, x = 'release_year', y = 'net_profit', scatter_kws = {'alpha' : .4},
           x_jitter = 0.5, y_jitter = 1, fit_reg = False)
```

自1960年起，电影净利润总体呈上升趋势。捞钱多的电影越来越多，但是亏损出现的几率也随之变大。不管在哪个年头，依然不乏有很多影片赚得一般般。
